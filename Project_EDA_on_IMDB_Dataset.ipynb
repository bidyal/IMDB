# 4.Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
imdb1
```

```python
imdb1.info()
```

```python
##imdb_crew=imdb1[imdb1['crew']!=imdb1['crew'].isna()]
imdb_crew=imdb1[~imdb1['crew'].isna()]
```

```python
imdb_crew.info()
```

```python
def get_prod_dir_list(x):
    try:
        pl=[]
        dl=[]
        for i in x:
            if i['job']=='Producer':
                pl.append(i['name'])
            if i['job']=='Director':
                dl.append(i['name'])
        return pl,dl
    except:
        return [],[]
```

```python
get_prod_dir_list(imdb_crew.loc[0,'crew'])
```

```python
imdb_crew[['Producer','Director']]=imdb_crew.apply(lambda x:get_prod_dir_list(x['crew']),axis=1,result_type='expand')
```

```python
imdb_crew
```

```python
imdb_crew=imdb_crew.sort_values(['ROI'],ascending=False)
```

```python
imdb_crew
```

# 5.Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
imdb_cast=imdb1[~imdb1['cast'].isna()]
```

```python
imdb_cast.info()
```

```python
imdb_cast.loc[0,'cast']
```

```python
def get_actors(x):
    try:
        a=[]
        for i in x:
            a.append(i['name'])
        return a
    except:
        return []
        
        
```

```python
get_actors(imdb_cast.loc[0,'cast'])
```

```python
imdb_cast['Actor_list']=imdb_cast['cast'].apply(get_actors)
```

```python
imdb_cast
```

```python
a_tab=[]
for i in imdb_cast['Actor_list']:
    for j in i:
        a_tab.append(j)
```

```python
a_table=pd.DataFrame(a_tab,columns=['actors'])
```
